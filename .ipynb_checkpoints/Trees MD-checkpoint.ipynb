# Susceptibility of Tree Species in Maryland to Climate Change per. Avg. of Three Climate Models (Hadley, PCM & GFDL)

The data imported is from the US Forestry Service Climate Change Atlas. The study uses the data to predict the future migration patterns of tree species in the aftermath of a climatic shift. 



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
trees = pd.read_csv('MD_mean_fz_winlose.csv')
```

```python
tree_names = trees['SppSN']
bar_widths = trees['Gcm3AvgHiDif'].values
bar_positions = np.arange(134) + .75
tick_positions = range(1,134)
fig, ax = plt.subplots(figsize=(50, 40))
ax.barh(bar_positions,np.sort(bar_widths), .5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(tree_names, fontsize=15)
ax.set_ylabel("Tree Species", fontsize=30)
ax.set_xlabel("GCM Rating", fontsize=30)
ax.set_title("Tree Susceptibility to Climate Change in MD", fontsize=50)
plt.show()
```

```python
fig1, ax1 = plt.subplots()
ax1.hist(bar_widths, 40)
ax1.set_xlim(-8,8)
ax1.set_xlabel('GCM Threat Rating')
ax1.set_ylabel('Number of Species')
ax1.set_xticks(np.arange(-8, 8, step=1))
ax1.set_title("Frequency in Susceptibility Amongst Tree Species in MD")
```

# Which Species are Most at Risk?
<h4>Sorted by Genus</h4>

```python
def get_genus(col):
    col = col.str.split().str[0]
    return col

trees['Genus'] = get_genus(trees['SppSN'])
genus_list = trees['Genus'].unique()
print(len(genus_list))

fig = plt.figure(figsize=(5,300))
counter = 1
for genus in genus_list:
    
    selected_rows = trees[trees['Genus'] == genus]
    #print(selected_rows[['Gcm3AvgHiDif', 'SppCN']])
    ax = fig.add_subplot(51, 1, counter)
    ax = selected_rows['Gcm3AvgHiDif'].plot(kind='hist', bins=range(-7,7))
    ax.set_title(genus)
    ax.set_xlabel('GCM3 Threat Rating')
    ax.set_ylabel('Number of Species')
    fig.subplots_adjust(hspace = .4)
    ax.set_xticks(np.arange(-7, 7, step=1))
    ax.set_yticks(np.arange(0, 20, step=1))
    counter += 1

```
